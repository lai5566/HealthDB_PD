## Health-01 Health-07 資料路徑與檔名

```python
import os
import pandas as pd
from sas7bdat import SAS7BDAT

# 指定目錄路徑
directory_Health_01_path = 'data/Health-01/Health-01'
directory_Health_07_path = 'data/Health-07'

# 列出資料夾下所有檔名
folders_Health_01 = [name for name in os.listdir(directory_Health_01_path) if
                     os.path.isfile(os.path.join(directory_Health_01_path, name))]
folders_Health_07 = [name for name in os.listdir(directory_Health_07_path) if
                     os.path.isfile(os.path.join(directory_Health_07_path, name))]
# 輸出結果
print(folders_Health_01)
countHealth_01 = len(folders_Health_01)
print("Health_01_file have: %d" % countHealth_01)

print("\n")
print(folders_Health_07)
countHealth_07 = len(folders_Health_07)
print("Health_07_file have: %d" % countHealth_07)


```

##  Health-01多sas檔合併df

```python
directory_Health_01_path = 'data/Health-01/Health-01'

df_mergedH_01 = pd.DataFrame()
totoalRowCount_H01 = 0
# 多檔案合併df
for fileName in folders_Health_01:
    #print(folders_Health_01)
    # print(f"{directory_Health_01_path}/{Filename}")
    # print("%s/%s"%(directory_Health_01_path,Filename))
    df_ReadH1 = SAS7BDAT(f"{directory_Health_01_path}/{fileName}").to_data_frame()
    #df_ReadH1=SAS7BDAT(("%d/%d"%(directory_Health_01_path,Filename))).to_data_frame()
    nowRead = len(df_ReadH1)
    totoalRowCount_H01 += nowRead
    print("%s_nowRead=%d,NowtotoalReadSum=%d" % (fileName, nowRead, totoalRowCount_H01))
    df_mergedH_01 = pd.concat([df_mergedH_01, df_ReadH1], axis=0)
    print("after merged df_mergedH_01 row count %d\n" % len(df_mergedH_01))
```

```python
display(df_mergedH_01)
```

## Health-07 多sas檔案合併

```python
directory_Health_07_path = 'data/Health-07'
df_mergedH_07 = pd.DataFrame()
totoalRowCount_H07 = 0
# 多檔案合併df
for fileName in folders_Health_07:
    #print(folders_Health_07)
    # print(f"{directory_Health_07_path}/{Filename}")
    # print("%s/%s"%(directory_Health_07_path,Filename))
    df_ReadH7 = SAS7BDAT(f"{directory_Health_07_path}/{fileName}").to_data_frame()
    #df_ReadH7=SAS7BDAT(("%d/%d"%(directory_Health_07_path,Filename))).to_data_frame()
    nowRead = len(df_ReadH7)
    totoalRowCount_H07 += nowRead
    print("%s_nowRead=%d,NowtotoalReadSum=%d" % (fileName, nowRead, totoalRowCount_H07))
    df_mergedH_07 = pd.concat([df_mergedH_07, df_ReadH7], axis=0)
    print("after merged df_mergedH_07 row count %d\n" % len(df_mergedH_07))
```
