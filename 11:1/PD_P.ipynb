```python
#odds ratio語法
import pandas as pd
import numpy as np
from scipy.stats import fisher_exact
import statsmodels.api as sm
from statsmodels.stats.contingency_tables import Table2x2
from datetime import datetime
#合併PD與nPD
df_d1_PDandnPD=pd.concat([df_d1_DP_group,nPD_d1_age])
# 創建交叉表
contingency_table = pd.crosstab(df_d1_PDandnPD['PD'], df_d1_PDandnPD['disease2'])


print(contingency_table.shape)
# 計算odds ratio和p-value
odds_ratio, p_value = fisher_exact(contingency_table)

# 計算odds ratio的信賴區間
table = Table2x2(contingency_table.values)
confidence_interval = table.oddsratio_confint()

print(f'Odds Ratio: {odds_ratio}')
print(f'p-value: {p_value}')
print(f'Confidence Interval: {confidence_interval}')
if p_value < 0.05:
    print("\nfail")
else:
    print("\npass")
    
    nowtime = datetime.now().strftime("%Y%m%d_%H%M")
    pass_csvFileName=f'd1_pass_PDandnPD{nowtime}.csv'
    print(pass_csvFileName)    
    df_d1_PDandnPD.to_csv(f'data/{pass_csvFileName}', index=False)
```

```python
import pandas as pd
from scipy.stats import fisher_exact
from statsmodels.stats.contingency_tables import Table2x2
from datetime import datetime

# 假设 df_d1_PDandnPD 是您的合并数据集
df_d1_PDandnPD = pd.concat([df_d1_DP_group, nPD_d1_age])

# 存储结果的列表
results = []
sumP=0
# 遍历 disease1 到 disease7
for n in range(1, 8):
    disease_col = f'disease{n}'
    
    # 筛选出 diseaseN == 1 的样本
    # disease_data = df_d1_PDandnPD[df_d1_PDandnPD[disease_col] == 1]
    
    # 创建 PD 与当前疾病的交叉表
    contingency_table = pd.crosstab(df_d1_PDandnPD['PD'], df_d1_PDandnPD[disease_col])
    
    # 检查表格是否为 2x2
    if contingency_table.shape == (2, 2):
        # 计算 Odds Ratio 和 p 值
        odds_ratio, p_value = fisher_exact(contingency_table)
        
        # 计算置信区间
        table = Table2x2(contingency_table.values)
        confidence_interval = table.oddsratio_confint()
        
        # 记录结果
        results.append({
            'Disease': disease_col,
            'Odds Ratio': odds_ratio,
            'p-value': p_value,
            'Confidence Interval': confidence_interval
        })
        
        # 输出结果
        print(f'\n{disease_col}:')
        print(f'Odds Ratio: {odds_ratio}')
        print(f'p-value: {p_value}')
        print(f'Confidence Interval: {confidence_interval}')
        
        # 检查 p 值
        if p_value < 0.05:
            print("Y_结果显著，存在统计学关联。")
        else:
            print("N_结果不显著，未发现统计学关联。")
            sumP+=1
    else:
        print(f'\n{disease_col}:')
        print("X_交叉表不是 2x2，无法计算统计量。请检查数据。")
print(sumP)
# 将结果转换为 DataFrame
# results_df = pd.DataFrame(results)
# 
# # 显示所有结果
# print("\n所有疾病的分析结果：")
# print(results_df)

# 保存结果到 CSV 文件
#nowtime = datetime.now().strftime("%Y%m%d_%H%M")
#results_df.to_csv(f'data/disease_analysis_results_{nowtime}.csv', index=False)
```
